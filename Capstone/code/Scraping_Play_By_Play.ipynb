###  Make request to get play by play for every second of NBA game.

```python
def parse_game(urls):   
    records = []
    for game in urls:
        print game
        r = requests.get('http://stats.nba.com/stats/winprobabilitypbp?GameID='+game +'&RunType=each+second', headers={"user-agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.92 Safari/537.36"})
        print r.url
        print r.status_code
        data = r.json()
        rows = data['resultSets'][0]['rowSet']
        for row in rows:
            records.append(row)
    return records
```

# WHOLE_GAME_FEATURE

I want to calculate total game metrics using it on play by play data and testing it on the boxscore. Once I verified that this worked, I was able to generate information for both halftime and quarter information.

```python
list_game_id = []
list_home_fg_made = []
list_home_fg_misses=[]
list_home_fg_pct =[]
list_visiting_fg_made=[] 
list_visiting_fg_misses=[] 
list_visiting_fg_pct=[]
list_home_rebounds=[] 
list_visiting_rebounds=[]

for game in nba_game_ids :
        df = nba[nba['GAME_ID']==game]
        visiting_fg_misses = len(df[df['DESCRIPTION'].str.contains('MISS') & ~df['DESCRIPTION'].str.contains('Free Throw') & (df['LOCATION']=='v') ] )
        home_fg_misses = len(df[df['DESCRIPTION'].str.contains('MISS') & ~df['DESCRIPTION'].str.contains('Free Throw') & (df['LOCATION']=='h') ] )
        visiting_fg_made = len(df[~df['DESCRIPTION'].str.contains('MISS') &  (df['LOCATION']=='v') &(df['DESCRIPTION'].str.contains('Layup') | df['DESCRIPTION'].str.contains('Dunk') | df['DESCRIPTION'].str.contains('Jump Shot') | df['DESCRIPTION'].str.contains('Hook Shot') | df['DESCRIPTION'].str.contains('Jumper') ) ]) 
        home_fg_made = len(df[~df['DESCRIPTION'].str.contains('MISS') &  (df['LOCATION']=='h') &(df['DESCRIPTION'].str.contains('Layup') | df['DESCRIPTION'].str.contains('Dunk') | df['DESCRIPTION'].str.contains('Jump Shot') | df['DESCRIPTION'].str.contains('Hook Shot')  ) ]) 
        visiting_fg_made, visiting_fg_misses+visiting_fg_made,home_fg_made,home_fg_misses + home_fg_made
        visiting_rebounds = len(df[(df['LOCATION']=='v') & df['DESCRIPTION'].str.contains('REBOUND')])
        home_rebounds = len(df[(df['LOCATION']=='h') & df['DESCRIPTION'].str.contains('REBOUND')])
        visiting_fg_pct = visiting_fg_made/float(visiting_fg_made+visiting_fg_misses)
        home_fg_pct = home_fg_made/float(home_fg_made+home_fg_misses)


       
        list_game_id.append(game)
        list_home_fg_made.append(home_fg_made)
        list_home_fg_misses.append(home_fg_misses)
        list_home_fg_pct.append(home_fg_pct)
        list_visiting_fg_made.append (visiting_fg_made)
        list_visiting_fg_misses.append(visiting_fg_misses)
        list_visiting_fg_pct.append(visiting_fg_pct)
        list_home_rebounds.append(home_rebounds)
        list_visiting_rebounds.append(visiting_rebounds)


full_time = pd.DataFrame( {'GAME_ID':list_game_id,
               'HOME_FG':list_home_fg_made,
               'HOME_FG_MISSES':list_home_fg_misses,
               'HOME_FG_PCT':list_home_fg_pct,
               'VISITING_FG':list_visiting_fg_made,
               'VISITING_MISSES':list_visiting_fg_misses,
               'VISITING_FG_PCT':list_visiting_fg_pct,
               'HOME_REBOUNDS':list_home_rebounds,
               'VISITING_REBOUNDS':list_visiting_rebounds

    
})

full_time.head()
        
        

   
```
